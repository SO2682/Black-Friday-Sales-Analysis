##Analysing Gender


```python
da=df.groupby('Gender').size()
ds=pd.DataFrame(da)
ds.rename({0:"Ratio"},axis=1,inplace=True)
ds.plot.pie(y='Ratio', figsize=(4,4) ,autopct= "%0.1f")

```

```python
df.groupby('Gender').sum()['Purchase'].plot(kind='pie',autopct="%0.2f" ,title='Purchases by gender')
df.groupby('Gender').sum()['Purchase']
```

```python
df.groupby('Gender').mean()['Purchase'].plot(kind='pie', autopct='%0.2f',title= 'Average Purchases by Gender')
```

## AGE AND MARITAL STATUS ANALYSIS


```python
df.groupby('Age').size().plot(kind='bar', title='Age wise Purchases')
```

```python
lst=[]
for i in df['Age'].unique():
      lst.append([i, df[df['Age'] == i ]['Product_ID'].nunique()])
unique = pd.DataFrame(lst, columns=['Age','Unique_Products'])
print(unique)
unique.plot(kind='bar', title= 'Unique products per age category', x='Age')
```

```python
df.groupby('Age').mean()['Purchase'].plot(kind='pie', autopct="%0.1f")
```
